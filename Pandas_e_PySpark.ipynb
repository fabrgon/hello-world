* **Filtrando dataframes**

Imagine que você quer trabalhar apenas com um conjunto específico do seu dataframe. Vamos supor que você queira trabalhar com o pedaço do dataframe onde o comprimento da sépala é maior ou igual a 7 cm. Isso é muito simples em pandas:

```python
df_filtrado = df[df['sepal length (cm)']>=7] #Esse filtro irá nos gerar um outro dataframe, por isso estou salvando ele com outro nome
```

```python
df_filtrado.reset_index()
```

Vamos ver se funcionou? Vamos usar o value_counts():

```python
df_filtrado['sepal length (cm)'].value_counts() #Funcionou! Só temos valores maiores ou iguais a 7 :)
```

Vamos supor agora que você queira trabalhar com um pedaço do dataframe onde o comprimento da sépala é maior ou igual a 7 cm e a largura da sépala é maior que 3.

```python
df_filtrado = df[(df['sepal length (cm)']>=7) & (df['sepal width (cm)']> 3)] #&, |
```

Vamos ver se funcionou? Vamos usar o value_counts():

```python
df_filtrado['sepal length (cm)'].value_counts() #Funcionou! Só temos valores maiores ou iguais a 7.
```

```python
df_filtrado['sepal width (cm)'].value_counts() #Funcionou! Só temos valores maiores que 3.
```

Agora vamos supor que você só quer estudar as Species iguais a 0.0. Podemos fazer o seguinte:

* **Juntando dataframes**

Para juntarmos dataframes, usamos o merge.

No código abaixo, estamos juntando dois dataframes com um left join (how = 'left'), usando como chave a coluna 'Species'.

* **Fazendo uma amostra do dataframe**

Fazemos isso com o .sample(). Ele nos retorna N linhas aleatórias do dataframe.

No código abaixo, n = 5 significa a quantidade de linhas a seres escolhidas, axis = 0 signifca que você quer uma amostra das linhas, random_state = 10 significa que se você rodar o código novamente virão as mesmas 5 linhas.

A correlação do Pandas é usada para determinar a correlação (ou seja, dependência) entre todas as colunas do dataframe. Os valores ausentes são excluídos e as colunas não numéricas também são ignoradas.

```python
data = [(1, 1), (2, 4), (3, 9), (4, 16), (5, 25)]
df = spark.createDataFrame(data, ["actual", "predicted"])

df.show()
```

```python
from pyspark.sql.functions import col
df = df.withColumn("squared_error", pow((col("actual") - col("predicted")), 2))

# Calculate the mean squared error
mse = df.agg({"squared_error": "avg"}).collect()[0][0]

print(f"Mean Squared Error (MSE) = {mse}")
```
