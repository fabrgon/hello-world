* **Filtrando dataframes**

Imagine que voc quer trabalhar apenas com um conjunto especfico do seu dataframe. Vamos supor que voc queira trabalhar com o pedao do dataframe onde o comprimento da spala  maior ou igual a 7 cm. Isso  muito simples em pandas:

```python
df_filtrado = df[df['sepal length (cm)']>=7] #Esse filtro ir nos gerar um outro dataframe, por isso estou salvando ele com outro nome
```

```python
df_filtrado.reset_index()
```

Vamos ver se funcionou? Vamos usar o value_counts():

```python
df_filtrado['sepal length (cm)'].value_counts() #Funcionou! S temos valores maiores ou iguais a 7 :)
```

Vamos supor agora que voc queira trabalhar com um pedao do dataframe onde o comprimento da spala  maior ou igual a 7 cm e a largura da spala  maior que 3.

```python
df_filtrado = df[(df['sepal length (cm)']>=7) & (df['sepal width (cm)']> 3)] #&, |
```

Vamos ver se funcionou? Vamos usar o value_counts():

```python
df_filtrado['sepal width (cm)'].value_counts() #Funcionou! S temos valores maiores que 3.
```

```python
data = [(1, 1), (2, 4), (3, 9), (4, 16), (5, 25)]
df = spark.createDataFrame(data, ["actual", "predicted"])

df.show()
```

```python
from pyspark.sql.functions import col
df = df.withColumn("squared_error", pow((col("actual") - col("predicted")), 2))

# Calculate the mean squared error
mse = df.agg({"squared_error": "avg"}).collect()[0][0]

print(f"Mean Squared Error (MSE) = {mse}")
```
